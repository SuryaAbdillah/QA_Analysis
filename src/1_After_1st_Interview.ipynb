## Feature engineering: known unknown competitor price

```python
df['is_compe_a'] = df['Competitor A'].notna()
df['is_compe_b'] = df['Competitor B'].notna()
df['is_compe_c'] = df['Competitor C'].notna()

df['known_num_compe'] = df[['is_compe_a', 'is_compe_b', 'is_compe_c']].sum(axis=1)
df
```

```python
df['avg_compe_price'] = df['Competitor A'] + df['Competitor B'] + df['Competitor C']
df['avg_compe_price'] = df['avg_compe_price'] / df['known_num_compe']

df['max_compe_price'] = df[['Competitor A', 'Competitor B', 'Competitor C']].max(axis=1)
df['min_compe_price'] = df[['Competitor A', 'Competitor B', 'Competitor C']].min(axis=1)

df
```

```python
for col in ['Competitor A', 'Competitor B', 'Competitor C']:
    print(f'\n=== {col} ===')

    result = (
        df.dropna(subset=[col])
          .groupby(['Kw', col, 'Product'], dropna=False)
          .size()
          .reset_index(name='count')
          .sort_values(['Kw', col, 'Product'])
    )

    display(result)
```

```python
for col in competitor_cols:
    print(f'\n=== Check consistency for {col} by Product + Kw ===')

    check_product_kw = (
        df.dropna(subset=[col])
          .groupby(['Product', 'Kw'])[col]
          .agg(
              unique_values=lambda x: sorted(x.unique()),
              n_unique='nunique',
              count='count'
          )
          .reset_index()
          .sort_values(['Kw', 'Product'])
    )

    display(check_product_kw)
```

```python
for col in competitor_cols:
    print(f'\n=== Imputation opportunity for {col} ===')

    temp = df.copy()

    # jumlah nilai non-null per Product + Kw
    temp[f'{col}_known_count'] = (
        temp.groupby(['Product', 'Kw'])[col]
            .transform(lambda x: x.notna().sum())
    )

    # jumlah unique non-null per Product + Kw
    temp[f'{col}_n_unique_known'] = (
        temp.groupby(['Product', 'Kw'])[col]
            .transform(lambda x: x.dropna().nunique())
    )

    opportunity = temp[
        temp[col].isna() &
        (temp[f'{col}_known_count'] > 0) &
        (temp[f'{col}_n_unique_known'] == 1)
    ][
        ['Quote ID', 'Product', 'Kw', col, f'{col}_known_count', f'{col}_n_unique_known']
    ]

    display(opportunity)
```

```python
for col in competitor_cols:
    print(f'\n=== {col}: consistency by Product ===')

    product_comp_check = (
        df.dropna(subset=[col])
          .groupby('Product')
          .agg(
              kw_unique=('Kw', lambda x: sorted(x.dropna().unique())),
              competitor_unique=(col, lambda x: sorted(x.dropna().unique())),
              n_unique_price=(col, lambda x: x.dropna().nunique()),
              count=(col, 'count')
          )
          .reset_index()
          .sort_values('Product')
    )

    display(product_comp_check)
```

```python
for col in competitor_cols:
    print(f'\n=== Missing check for known-product {col} ===')

    # Product yang punya harga competitor untuk kolom tersebut
    known_products = df.loc[df[col].notna(), 'Product'].unique()

    # Row dengan product tersebut, tapi competitor price masih kosong
    missing_known_product = df[
        df['Product'].isin(known_products) &
        df[col].isna()
    ][
        ['Quote ID', 'Product', 'Kw', 'Unit Price', 'Qty', 'Subtotal Price',
         'Competitor A', 'Competitor B', 'Competitor C']
    ].sort_values(['Product', 'Kw', 'Quote ID'])

    print(f'Number of rows that can potentially be imputed by Product: {len(missing_known_product)}')
    display(missing_known_product)
```

### oke gas oke gas buat fiturnya

```python
df_fe = (
    df.sort_values(['Quote ID', 'Product', 'Unit Price'])
      .assign(
          price_order=lambda x: x.groupby(['Quote ID', 'Product']).cumcount() + 1,
          is_highest_price=lambda x: (
              x['Unit Price'] == x.groupby(['Quote ID', 'Product'])['Unit Price'].transform('max')
          ).astype(int)
      )
)

same_quote_product = df_fe[
    df_fe.groupby(['Quote ID', 'Product'])['Quote ID']
         .transform('count') > 1
]

same_quote_product[
    ['Quote ID', 'Product', 'Kw', 'Unit Price',
     'price_order', 'is_highest_price',
     'Convert to Order (0:Success, 1:Fail)']
].sort_values(['Quote ID', 'Product', 'price_order'])
```

```python
same_quote_product = df_fe[
    df_fe.groupby(['Quote ID', 'Product'])['Quote ID']
         .transform('count') > 1
].copy()
```

#### success all

```python
success_all = same_quote_product[
    same_quote_product.groupby(['Quote ID', 'Product'])['Convert to Order (0:Success, 1:Fail)']
        .transform(lambda x: (x == 0).all())
].copy()

print(f"Number of rows that success all: {len(success_all)}")

success_all[
    ['Quote ID', 'Product', 'Kw', 'Unit Price',
     'price_order', 'is_highest_price',
     'Convert to Order (0:Success, 1:Fail)']
].sort_values(['Quote ID', 'Product', 'price_order'])
```

```python
# SAME UNIT PRICE: ini apa ditanyain ya, apakah ada kemungkinan dalam 1 quote id terdapat beberapa product id sama dan harga yang sama
success_all_same_price = success_all[
    success_all.groupby(['Quote ID', 'Product'])['Unit Price']
        .transform('nunique') == 1
].copy()

success_all_same_price[
    ['Quote ID', 'Product', 'Kw', 'Unit Price',
     'price_order', 'is_highest_price',
     'Convert to Order (0:Success, 1:Fail)']
].sort_values(['Quote ID', 'Product', 'price_order'])
```

```python
mixed_success_fail = same_quote_product[
    same_quote_product.groupby(['Quote ID', 'Product'])['Convert to Order (0:Success, 1:Fail)']
        .transform(lambda x: x.nunique() > 1)
].copy()

print(f"Number of rows that mixed success and fail: {len(mixed_success_fail)}")

mixed_success_fail[
    ['Quote ID', 'Product', 'Kw', 'Unit Price',
     'price_order', 'is_highest_price',
     'Convert to Order (0:Success, 1:Fail)']
].sort_values(['Quote ID', 'Product', 'price_order'])
```

## Feature enginerering: perhitungan

```python
df_fe['min_competitor_price'] = df_fe[competitor_cols].min(axis=1, skipna=True)

df_fe['price_gap_vs_min_competitor'] = (
    df_fe['Unit Price'] - df_fe['min_competitor_price']
)

df_fe['price_ratio_vs_min_competitor'] = (
    df_fe['Unit Price'] / df_fe['min_competitor_price']
)
```

```python
df_fe['has_energy_grant'] = df_fe['Energy grant amount'].notna().astype(int)

df_fe['effective_price_after_grant'] = (
    df_fe['Subtotal Price'] - df_fe['Energy grant amount']
)

df_fe['grant_ratio_to_subtotal'] = (
    df_fe['Energy grant amount'] / df_fe['Subtotal Price']
)
```

```python
df_fe['estimated_cost'] = (
    df_fe['Subtotal Price'] * (1 - df_fe['Gross Margin Rate'])
)

df_fe['estimated_gross_profit'] = (
    df_fe['Subtotal Price'] - df_fe['estimated_cost']
)
```
